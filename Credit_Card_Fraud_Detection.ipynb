## Importing Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

## Loading dataset

```python
cols = ['Time','V1','V2','V3','V4','V5','V6','V7','V8','V9','V10',
        'V11','V12','V13','V14','V15','V16','V17','V18','V19',
        'V20','V21','V22','V23','V24','V25','V26','V27','V28',
        'Amount','Class']

credit_card_data = pd.read_csv(
    "/content/creditcard.csv",
    on_bad_lines='skip',
    names=cols,
    header=0
)
credit_card_data
```

## As this dataset is highly unbalanced

```python
# seeparting fraud and real transaction for analysis
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
print(legit.shape)
print(fraud.shape)
```

```python
fraud.Amount.describe()
```

```python
## comparing the values for both transaction
credit_card_data.groupby('Class').mean()
```

```python
## building sample data set contaning simple distrbution
legit_sample=legit.sample(n=492)
```

```python
# concatenating both samplesa
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

## Resamplying because the data set is highly unbalanced

```python
from sklearn.utils import resample


# legit sample (already 492)
legit_sample = legit.sample(n=492, random_state=42)

# fraud ko 492 tak upsample karo
fraud_upsampled = resample(fraud,
                          replace=True,   # important
                          n_samples=492,
                          random_state=42)

# combine
new_dataset = pd.concat([legit_sample, fraud_upsampled])

# check
print(new_dataset['Class'].value_counts())
```

```python
a=new_dataset.groupby('Class').mean
a
```

## splitting the data in to featurres and target

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
X
```

```python
Y
```

## splitting in to training and tsting

```python
### fr traning
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
X.shape,X_train.shape,X_test.shape
```

## model training

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

## Model Evaluation

```python
# accuracy_score on training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print(f"Accuracy score on training data:",training_data_accuracy)
```

```python
## accuracy on test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print(f"Accuracy score on testing data:",test_data_accuracy)
```

# now making function to check the legiit  or fraud transaction of data

```python
def predict_transaction(input_data):
    prediction = model.predict([input_data])

    if prediction[0] == 0:
        return "Legit Transaction"
    else:
        return "Fraud Transaction"
```
